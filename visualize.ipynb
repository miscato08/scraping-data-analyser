```python
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
df = pd.read_csv('data_vs2.csv')
print(df.head())
```

```python
# use a style
plt.style.use("seaborn")
```

```python
# set up the figure and axes
fig, (ax0, ax1, ax2) = plt.subplots(1, 3)

# scatter plots
scatter0 = ax0.scatter(x=df["age"], y=df["rank"], cmap="winter")
scatter1 = ax1.scatter(x=df["years"], y=df["rank"], cmap="winter")
scatter2 = ax2.scatter(x=df["length"], y=df["rank"], cmap="winter")

# add labels and titles
ax0.set(title="Rankings with Age factor", ylabel="rank")
ax1.set(title="Rankings with Years factor", ylabel="rank")
ax2.set(title="Rankings with Length factor", ylabel="rank")

# slim the x axis
ax0.set_xlim([50, 80])
ax0.set_ylim([60, 200])
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax2.set_xlim([50, 80])
ax2.set_ylim([60, 200])

# create a legend
ax0.legend(*scatter0.legend_elements(), title="Target")

# Create a mean line
ax0.axhline(y=df["rank"].mean(), linestyle="--")
ax1.axhline(y=df["rank"].mean(), linestyle="--")
ax2.axhline(y=df["rank"].mean(), linestyle="--")

# add title for whole plot
fig.suptitle("Film Rankings Analysis", fontsize=16, fontweight="bold")

# show the plot
plt.show()
```

```python
# save it
fig.savefig("film-ranking-analysis.png")
```
